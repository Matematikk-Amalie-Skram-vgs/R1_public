## 1A Potenser
Regneopperatoren for opphøyd i er ** i python    
Under er potensen $2^4$ regnet ut.


```python
print(2**4)
```

Ved hjelp av en f-streng kan svaret presenteres bedre. Her er noen eksempler, legg merke til at det som står i krøllparantesen regnes ut:

```python
print(f'8^3 = {8**3}')
print(f'8^(1/2) = {8**(1/2)}')
print(f'8^(1/2) = {8**(1/2):.2f}') # formatert som flyttall, med 2 desimaler
print(f'42^34 = {42**34:.3e}') # Skriv på standardform med 3 desimaler.
```

## 1B Logaritmer 
Python har ikke logaritmer som innebygde funksjoner så disse må importeres. Logaritmer kan importeres fra flere bibliotek og på flere måter. Vi velger Numpy, siden dette er et bibliotek vi skal bruke til mye. Når vi bare skal bruke noen få er en mulighet å kun importere bare disse. Når vi lager litt større kode fortrekker vi å importere ved å lage aliaserer.   
**NB funksjonen trenger bare importes en gang pr. notatbok**

```python
from numpy import log10 
print(f'lg(500) = {log10(500)}')
```

### Oppgave 3
Skriv en programsnutt som regner ut logaritmene:

## Naturlige logaritmer 
I numpy heter heter funksjonen for naturlige logaritmer `log()`.
Legg til en kodelinje som importer `log()` slik at kodecellen under kjører uten feilmelding

```python
from numpy import log
print(f'ln(2) = {log(2)}')
```

### Oppgave 5
Skriv en kodesnutt som regner ut logarimene slik at du kan sammenligne:   
a) $\ln(4)$ med $2\cdot \ln(2)$   
b) $\ln(8)$ med $3\cdot \ln(2)$   
c) $\ln(16)$ med $4\cdot \ln(2)$   
d) $\ln(32)$ med $5\cdot \ln(2)$

```python
print(f'ln(4) = {log(4)} og 2*ln(2) = {2*log(2)}')
print(f'ln(8) = {log(8)} og 3*ln(2) = {3*log(2)}')
print(f'ln(16) = {log(16)} og 4*ln(2) = {4*log(2)}')
print(f'ln(32) = {log(32)} og 5*ln(2) = {5*log(2)}')
```
